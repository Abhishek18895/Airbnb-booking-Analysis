# **using IQR technique**

```python

# writing a outlier function for removing outliers in important columns.


def iqr_technique(DFcolumn):
    Q1 = np.percentile(DFcolumn, 25)
    Q3 = np.percentile(DFcolumn, 75)
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range

lower_bound, upper_bound = iqr_technique(Airbnb['price'])

Airbnb = Airbnb[(Airbnb['price'] > lower_bound) & (Airbnb['price'] < upper_bound)]

max_price = Airbnb['price'].max()
print(max_price)
```

## **What is the room count in overall NYC according to the listing of room types?**

```python
#plt.rcParams['figure.figsize'] = (8,5)
plt.figure(figsize=(12,6))
ax = sns.countplot(y='room_type', hue='neighbourhood_group',data=Airbnb,palette='bright')

total = len(Airbnb['room_type'])
for p in ax.patches:
                 percentage = '{:0.1f}%'.format(100 * p.get_width()/total)
                 x = p.get_x() + p.get_width() + 0.02
                 y = p.get_y() + p.get_height()/2
                 ax.annotate(percentage, (x,y))

plt.title('Count of each room type in NYC',fontweight='bold')
plt.xlabel('Rooms',fontweight='bold')
plt.xticks(rotation=90)
plt.show()
```

* Manhattan has more listed propeties with  Entire home/apt around 27% of total listed properties followed by Brooklyn with around 19.6%.

* Private room are more in Brooklyn as in 20.7% of the total listed property followed by with 16.3% of them while 6.9% private rooms from queens.

* We can infer that Brookly , Queens, Bronx has more private room types while Manhattan which has the higest number of listing in entire NYC has more Entire Home/apt room types.
